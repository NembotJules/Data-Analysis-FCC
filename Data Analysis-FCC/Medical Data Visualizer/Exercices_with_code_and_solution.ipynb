```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data
df = pd.read_csv('medical_examination.csv')
```

```python
df
```

```python
#Calculating the BMI
BMI = df.weight / (df.height*0.01)**2
# Adding the overweight column to the dataset, with values 0 = Not overweight, and 1= overweight
df['overweight'] = (BMI > 25).astype(int)
```

```python
df['overweight']
```

```python
#Normalize the data by making 0 always good and 1 always bad.
#If the value of cholesterol is 1, make the value 0. If the value is more than 1, make the value 1
df['cholesterol'] = np.where(df['cholesterol'] > 1, 1, 0)
```

```python
#Normalize the data by making 0 always good and 1 always bad.
#If the value of gluc is 1, make the value 0. If the value is more than 1, make the value 1
df['gluc'] = np.where(df['gluc'] > 1, 1, 0)
```

```python
df.head()
```

```python
#transform data into a long format, which makes it easier to analyze and visualize.
df_cat=df.melt(id_vars=['cardio'],value_vars=sorted(['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']))
```

```python
df_cat 
```

```python
#create a chart that shows the value counts of the categorical features using seaborn's catplot().
#The dataset should be split by 'Cardio' so there is one chart for each cardio value
sns.catplot(x='variable', kind='count', hue='value', data=df_cat, col = 'cardio')
```

```python
#Cleaning the data
df_heat=df[(df['ap_lo']<=df['ap_hi'])&(df['height']>=df['height'].quantile(0.025)) & (df['height']<=df['height'].quantile(0.975))
          &(df['weight']>=df['weight'].quantile(0.025)) & (df['weight']<=df['weight'].quantile(0.975))]
```

```python
#Creating the Correlation Matrix
corr_matrix=df_heat.corr()
corr_matrix
```

```python
#Creating the mask for the upper triangle of the correlation matrix
mask=np.triu(corr_matrix)
```

```python
mask
```

```python
#Plot the correlation matrix using seaborn's heatmap()
ax = sns.heatmap(
        corr_matrix, annot=True, fmt='.1f', linewidths=1, mask=mask, 
        vmax=.8, center=0.09, square=True, cbar_kws={'shrink':0.5})
```
